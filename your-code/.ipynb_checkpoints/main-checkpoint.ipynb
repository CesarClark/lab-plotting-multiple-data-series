# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import numpy as np
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
salesdf=data[['RetailSales', 'RetailTransfers', 'WarehouseSales', 'ItemType']].groupby(data.ItemType).agg('sum')


salesdf
```

```python
salesdf.plot.bar();
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
topsuppliers=data.copy()
topsuppliers['TotalSales']=topsuppliers.RetailSales + topsuppliers.RetailTransfers + topsuppliers.WarehouseSales
topsuppliers
```

```python
topsuppliers2=topsuppliers.groupby('Supplier').sum().sort_values(by='TotalSales', ascending=False)

topsuppliers2.TotalSales[0:9].plot.barh()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
#df['DATE'] = pd.to_datetime(df[['YEAR', 'MONTH']].assign(DAY=1))
#topsuppliers=topsuppliers.rename(columns={'Year':'YEAR', 'Month':'MONTH'})

data2=data.copy()
data2['full_date']=pd.to_datetime(data[['Year','Month']].assign(DAY=1))

multi_line=data2.groupby('full_date')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].mean()


multi_line[['RetailSales', 'RetailTransfers', 'WarehouseSales']].plot()
```

## 4. Plot the same information as above but as a bar chart.

```python
multi_line[['RetailSales', 'RetailTransfers', 'WarehouseSales']].plot.bar()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
data3=data2[['full_date', 'ItemType', 'RetailSales']]
data3
```

```python
#item_group=data2.groupby(['full_date']).sum()
item_group=data3.pivot_table(index=['full_date'], columns=['ItemType'], aggfunc=np.sum)
item_group
```

```python
item_group.plot()
plt.legend(bbox_to_anchor=(1, 1), loc=2) #This specirfic line is to make the legends stay away from the coordinate chart
```

## 6. Plot the same information as above but as a bar chart.

```python
item_group.plot.bar()
plt.legend(bbox_to_anchor=(1, 1), loc=2) 
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.scatterplot(x=data.RetailSales, y=data.RetailTransfers, hue=data.ItemType);
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.
